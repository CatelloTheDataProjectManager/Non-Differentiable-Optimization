**Calculate the derivative of the function**

In order to find the minimum of a function, we can calculate its derivative and set it equal to zero to find the critical points. At these critical points, the derivative is zero, which means that the slope of the function is zero. If the function is continuous and the derivative changes sign at a critical point, then that critical point is a local minimum.

**The first derivative**

The first derivative is commonly referred to as the "first derivative" in English. It measures the rate of change of a function with respect to its independent variable. In other words, it represents the slope of the tangent line to the curve of the function at a given point.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return (x-1)**2 * (x+2)

# Define the derivative of the function
def f_prime(x):
    return 3*x**2 - 2*x - 3

# Generate x values
x = np.linspace(-5, 5, 400)

# Generate y values for the function and its derivative
y = f(x)
y_prime = f_prime(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x-1)^2 * (x+2)')
plt.plot(x, y_prime, label="f'(x) = 3x^2 - 2x - 3")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of f(x) and its derivative')
plt.grid(True)
plt.legend()
plt.show()
```

**Using the derivative, find the critical points of the function.**

In mathematics, critical points of a function are the points where the derivative is equal to zero or undefined. These points are significant because they can indicate local extrema (minima or maxima) or points of inflection of the function. In the context of the function f(x) = (x-1)^2 * (x+2), the critical points are the solutions to the equation f'(x) = 0.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root

# Define the function
def f(x):
    return (x-1)**2 * (x+2)

# Define the derivative of the function
def f_prime(x):
    return 3*x**2 - 2*x - 3

# Generate x values
x = np.linspace(-5, 5, 400)

# Generate y values for the function
y = f(x)

# Find the critical points using scipy.optimize.root
critical_points = root(f_prime, [-1, 3])

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x-1)^2 * (x+2)')
plt.plot(critical_points.x, f(critical_points.x), 'ro', label='Critical points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = (x-1)^2 * (x+2)')
plt.grid(True)
plt.legend()
plt.show()

critical_points.x
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variables
x, y = sp.symbols('x y')

# Define the function
f = x**2 + y**2

# Calculate the gradient
gradient_f = [sp.diff(f, var) for var in (x, y)]

# Convert the gradient to a numerical function
gradient_func = sp.lambdify((x, y), gradient_f)

# Gradient Descent Parameters
learning_rate = 0.1  # Step size
max_iterations = 100  # Maximum number of iterations
tolerance = 1e-6  # Tolerance for stopping criterion

# Initial point
point = np.array([5.0, 5.0])  # Starting point (x, y)

# To store points for visualization
points = [point.copy()]

# Gradient Descent Loop
for iteration in range(max_iterations):
    grad = np.array(gradient_func(*point), dtype=float)  # Calculate gradient at current point
    new_point = point - learning_rate * grad  # Update point

    # Check for convergence (if the change is small)
    if np.linalg.norm(new_point - point) < tolerance:
        print(f"Converged in {iteration} iterations.")
        break
    
    point = new_point  # Move to the new point
    points.append(point.copy())  # Store the new point for visualization

print(f"Minimum found at: {point}")
print(f"Minimum value of f at this point: {f.subs({x: point[0], y: point[1]})}")

# Visualization
# Create a grid of (x, y) points
x_vals = np.linspace(-6, 6, 100)
y_vals = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X**2 + Y**2  # Calculate the function values

# Create the contour plot
plt.figure(figsize=(10, 6))
contour = plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8)

# Plot the gradient descent path
points = np.array(points)  # Convert list of points to a numpy array
plt.plot(points[:, 0], points[:, 1], marker='o', color='red', label='Gradient Descent Path')

# Mark the minimum point
plt.plot(point[0], point[1], marker='x', color='blue', markersize=10, label='Minimum Point (0, 0)')

# Add labels and legend
plt.title('Gradient Descent on $f(x, y) = x^2 + y^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.colorbar(label='Function Value')
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Step 1: Define the variable
x = sp.symbols('x')

# Step 2: Define the function
f = x**2 + 4*x + 4

# Step 3: Calculate the first derivative
f_prime = sp.diff(f, x)

# Step 4: Find the critical points by setting the first derivative to zero
critical_points = sp.solve(f_prime, x)

# Step 5: Calculate the second derivative
f_double_prime = sp.diff(f_prime, x)

# Step 6: Evaluate the second derivative at the critical point
second_derivative_value = f_double_prime.subs(x, critical_points[0])

# Step 7: Determine the nature of the critical point
if second_derivative_value > 0:
    nature = "local minimum"
elif second_derivative_value < 0:
    nature = "local maximum"
else:
    nature = "inconclusive"

# Display the results
print("The critical points are:", critical_points)
print("The second derivative at the critical point is:", second_derivative_value)
print("The critical point is a:", nature)

# Step 8: Plotting the function
# Create a range of x values
x_vals = np.linspace(-5, 2, 100)
# Calculate the corresponding y values for the function
f_vals = [f.subs(x, val) for val in x_vals]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = x^2 + 4x + 4', color='blue')
plt.scatter(float(critical_points[0]), float(f.subs(x, critical_points[0])), color='red', label='Critical Point (-2, 0)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Plot of the Function with Critical Point')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

## Minimization of a Cubic Function

1. Determine the gradient of the function.
2. Identify the critical points (the points where the gradient is zero).
3. Calculate the Hessian matrix.
4. Determine the nature of the critical points:
5. Minimum
6. Maximum
7. Saddle point

```python
import numpy as np
import sympy as sp

# Define the variables
x, y = sp.symbols('x y')

# Define the function
f = x**3 - 6*x**2 + 9*x + y**3 - 6*y

# Step 1: Gradient
grad_f = [sp.diff(f, var) for var in (x, y)]

# Step 2: Solve for critical points (where gradient is zero)
critical_points = sp.solve(grad_f, (x, y))

# Step 3: Hessian matrix
Hessian = sp.hessian(f, (x, y))

# Step 4: Determine the nature of the critical points
nature_of_points = []
for point in critical_points:
    H_val = Hessian.subs({x: point[0], y: point[1]})
    det_H = H_val.det()
    
    if det_H > 0:
        if H_val[0, 0] > 0:
            nature_of_points.append((point, 'Local Minimum'))
        else:
            nature_of_points.append((point, 'Local Maximum'))
    elif det_H < 0:
        nature_of_points.append((point, 'Saddle Point'))
    else:
        nature_of_points.append((point, 'Inconclusive'))

# Display results
grad_f, critical_points, Hessian, nature_of_points
```

**The gradient descent algorithm VS minimize_scalar from scipy.optimize**

```python
import numpy as np
from scipy.optimize import minimize_scalar

# Define the function f(x)
def f(x):
    return x**4 - 10*x**2 + 9

# Define the gradient f'(x)
def gradient(x):
    return 4*x**3 - 20*x

# Implement the gradient descent algorithm
def gradient_descent(starting_point, learning_rate=0.01, tolerance=1e-6, max_iterations=10000):
    x = starting_point
    for _ in range(max_iterations):
        grad = gradient(x)
        x_new = x - learning_rate * grad
        
        # Check for convergence
        if abs(x_new - x) < tolerance:
            break
        x = x_new
        
    return x

# Apply gradient descent starting from a smaller x0
x0 = 2  # Changed starting point to 2
learning_rate = 0.01  # Reduced learning rate
minimum_gd = gradient_descent(x0, learning_rate)

# Use minimize_scalar from scipy.optimize for comparison
result_scipy = minimize_scalar(f)

minimum_gd, result_scipy.x, f(minimum_gd), f(result_scipy.x)
```

## Fitting an exponential curve

We have a set of data (x, y) representing the evolution of a population over time. We want to fit an exponential model of the form to this data.
Use the minimize function from scipy.optimize to find the parameters that minimize the mean squared error between the values predicted by the model and the observed values.

**y = a exp(bx)**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
n_points = 1000

# Define the constants
a = 2
b = 0.5

# Generate time values
t = np.random.rand(n_points) * 10

# Increase the noise standard deviation to 20
noise_std = 20

# Calculate the population values with added Gaussian noise
population = a * np.exp(b * t) + np.random.randn(n_points) * noise_std

# Scale and shift the population values to fit within the range [0, 1000]
population = (population - np.min(population)) / (np.max(population) - np.min(population)) * 900 + 100

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(t, population, label='Noisy points')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Evolution of a population over time with added noise')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({'time': t, 'population': population})

df.info()
```

## Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np

def linear_regression_analytical(x, y):
    N = len(x)
    m = (N * np.sum(x * y) - np.sum(x) * np.sum(y)) / (N * np.sum(x**2) - np.sum(x)**2)
    b = (np.sum(y) - m * np.sum(x)) / N
    return m, b


# Calculate the slope and y-intercept of the best fit line
x = df['time']
y = df['population']
m_analytical, b_analytical = linear_regression_analytical(x, y)

# Create a scatter plot of the data
plt.scatter(x, y, color='blue')

# Create a line using the slope and y-intercept
plt.plot(x, m_analytical*x + b_analytical, color='red')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Linear Regression: Time vs Population')

# Show the plot
plt.show()
```

## Mean Squared Error (MSE)

```python
import numpy as np
import matplotlib.pyplot as plt

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def linear_regression_gradient_descent(x, y, learning_rate=0.01, n_iterations=10):
    m, b = 0, 0  # Initialization
    N = len(y)

    # History for visualization
    mse_history = []
    m_history, b_history = [], []

    for _ in range(n_iterations):
        y_pred = m * x + b
        dm = (-2/N) * np.sum(x * (y - y_pred))  # Derivative with respect to m
        db = (-2/N) * np.sum(y - y_pred)        # Derivative with respect to b
        m -= learning_rate * dm
        b -= learning_rate * db

        # Record history for visualization
        mse_history.append(mean_squared_error(y, y_pred))
        m_history.append(m)
        b_history.append(b)

    return m, b, mse_history, m_history, b_history

# Define the input and output data
x = df['time']
y = df['population']

# Calculate the slope, y-intercept, and MSE history using gradient descent
m_gd, b_gd, mse_history, m_history, b_history = linear_regression_gradient_descent(x, y)

# Create a plot of the MSE history
plt.plot(range(len(mse_history)), mse_history)

# Add labels and title
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.title('Convergence of Gradient Descent')

# Show the plot
plt.show()
```
